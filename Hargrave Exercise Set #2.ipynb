Consider the function $f(x) = x e^x$

### Part a:

##### Section 1

Use forward-difference algorithm to calculate an approximation to calculate $f''(2)$ for $h=0.5$, $h=0.45$, ... $h=0.05$.

The forward-difference approximation states that  $f'(x) \approx \frac{f(x+h) - f(x)}{h}$ and $f'(x+h) \approx \frac{f(x+2h) - f(x+h)}{h}$

Using the same approximation we find: $f''(x) \approx \frac{f'(x+h) - f'(x)}{h}$

which becomes $$f''(x) \approx \frac{f(x+2h) - 2f(x+h) + f(x)}{h^2}$$

```python
import numpy as np #import packages

#define a funciton that returns the avalue of f(x)
def f(x):
    return x * np.exp(x)

# create an array that contains the values 0.5, 0.45, ..., 0.05
h = np.arange(0.5, 0.0, -0.05)

# set x so that the derivative can be evaluated at some particular point
x=2

# approximate the second derivative of f(x) evaluated at x=2 using the forward difference approximation for the various values of h
for i in range (0, len(h)):

    second_derivative = (f(x+2*h[i]) - 2* f(x+h[i]) + f(x))/(h[i]**2)

    #print the calculated values of the second derivative at x=2
    print ('\n \n Using h = %.2f, the second derrivative of x * e^x evaluated at x=2 is approximated to be: \n %.15f' % (h[i], second_derivative))
```

#### Section 2

Use the central-difference approximation to calculate $f''(2)$ for $h=0.5$, $h=0.45$, ... $h=0.05$.

The central-difference approximation states that  $f'(x+h/2) \approx \frac{f(x+h) - f(x)}{h}$ and $f'(x-h/2) \approx \frac{f(x) - f(x-h)}{h}$

Using the same approximation we find: $f''(x) \approx \frac{f'(x+h/2) - f'(x-h/2)}{h}$

which becomes $$f''(x) \approx \frac{f(x+h) - 2f(x) + f(x-h)}{h^2}$$

This should look familiar from Excercise Set 1


```python
import numpy as np # import packages

# define a function that returns the value of f(x)
def f(x):
    return x * np.exp(x)

# create an array that contains the values 0.5, 0.45, ..., 0.05
h = np.arange(0.5, 0.0, -0.05)

# set x so that the derivative can be evaluated at some particular point
x=2

# approximate the second derivative of f(x) evaluated at x=2 using the central difference approximation for the various values of h
for i in range (0, len(h)):

    second_derivative = (f(x+h[i]) - 2 * f(x) + f(x-h[i]))/(h[i]**2)

    # print the calculated values of the second derivative at x=2
    print ('\n \n Using h = %.2f, the second derrivative of x * e^x evaluated at x=2 is approximated to be: \n %.15f' % (h[i], second_derivative))
```

## Excercise 2: Calculations with the relativistic Breit-Wigner function

The Breit-Wigner probability function is defined to be:

$$f(E) = \frac{k}{(E^2 - M^2)^2 + M^2 \Gamma^2}$$

where:

$$ k = \frac{2 \sqrt{2} M \Gamma \gamma}{\pi \sqrt{M^2 + \gamma}}$$

and 

$$ \gamma = \sqrt{M^2(M^2+\Gamma^2)} $$

### Part a:

Plot the Breit-Wigner distrobution with $M = 90 \mathrm{GeV}$ and $\Gamma = 10 \mathrm{GeV}$

```python
# import libraries
import numpy as np

# set values of relevant variables
M = 90
Gamma = 10
h = 1 #step width

# declare array that will contain the values of f(E) for various values of E
Breit_Wigner_array=[]

# define gamma in the program
gamma = np.sqrt(M**2 * (M**2 + Gamma**2))

# define k in the program
k = (2 * np.sqrt(2) * M * Gamma * gamma) / (np.pi * np.sqrt(M**2 + gamma))

# define function that returns value of f(E)
def f(E):
    return k / ((E**2 - M**2)**2 + M**2 * Gamma**2)

# create an array that contains values of h ranging between 0 and 180
E = np.arange(0,180 + h, h)

# for every value of E in the array E, find and store the value of f(E)
for i in range (0, len(E)):
    Breit_Wigner_array.append(f(E[i]))
    
# Convert arrays to vectors
E_vector = np.asarray(E)
Breit_Wigner_vector = np.asarray(Breit_Wigner_array)

# Plot f(E) against E    
plt.plot(E_vector,Breit_Wigner_vector)
plt.title('Breit-Wigner Probaiblity Function')
plt.ylabel('$f(E)$')
plt.xlabel('$E \; (\mathrm{GeV})$')
plt.show()
```

```python
# import packages
import numpy as np
import scipy.integrate as sp

# define relevant variables
M = 90
Gamma = 10
N = 181 # number points at which f(E) will be evaluated
a = 0 # lower bound of integration
b = 180 # upper bound of integration
h = (b - a) / (N - 1) # step size

# simpson's rule requires an even number of sections, returns error if this is not the case
if (N-1) % 2 == 1:
    print ('Error: You have divided the function into an odd number of intervals. Increase N by 1 and try again')

else:    
    # define gamma for the program
    gamma = np.sqrt(M**2 * (M**2 + Gamma**2))
    
    # define k for the program
    k = (2 * np.sqrt(2) * M * Gamma * gamma) / (np.pi * np.sqrt(M**2 + gamma))
    
    # define function that returns the value of f(E)
    def f(E):
        return k / ((E**2 - M**2)**2 + M**2 * Gamma**2)

    x = np.arange(a, b + h, h)     # create array that contains various values of E at which f(E) will be evaluated
    y = [] # declares array that will contain the various values of f(E) evaluated at the values of E stored in x

    # fill y array with the values of f(E) evaluated at the points E stored in array x
    for i in range (0, len(x)):
        y.append(f(x[i]))
    
    #defines integral variable that will be summed itteratively
    integral = 0

    # performs simpson integration
    for i in range (0, len(x) - 1):
        if i == 1 or i == len(x) - 1:
            integral += (1/3) * h * y[i]
        if i != 1 and i != len(x) - 1 and i % 2 == 0:
            integral += (4/3) * h * y[i]
        if i != 1 and i != len(x) -1 and i % 2 == 1:
            integral += (2/3) * h * y[i]
    
    # calls pre-existing function to perform simpson integration (as found in the scipy library)
    simps = sp.simps(y, x, h)
    
    # calcualte the agreement between the calculated value and the value returned by the sp.simps
    agreement = (1 - np.abs((integral-simps)/simps)) * 100

    #print the calculated value, the "true" value (according to sp.simps) and the agreement between the two
    print ('\n My program calculates the integral of the Breit-Wigner function to be %.15f' % (integral))

    print ('\n Scipy\'s .simps function calculates the integral of the Breit-Wigner function to be %.15f' % (simps))

    print('\n My program and the .simps function are in %.13f percent agreement' % (agreement))

```

### Diffraction at Knife's Edge

The intesity of light diffracted around a straight edge is:

$$ I = \frac{I_0}{2} \{[C(v) + 0.5]^2 + [S(v) + 0.5]^2 \}$$

where:

$$ C(v) = \int_0^v \cos{\pi w^2 / 2} dw$$

and

$$ S(v) = \int_0^v \sin{\pi w^2 / 2} dw$$

### Part a

Evaluate $I/I_0$ as a function of $v$

```python
#import libraries
import numpy as np
import scipy.integrate as sp

#define relevant variables
a = 0
N = 11

# define function that returns the value of the integrand of C evaluated at w
def c(w):
    return np.cos((np.pi * w**2)/2)

# define function that returns the value of the integrand of S evaluated at w
def s(w):
    return np.sin((np.pi * w**2)/2)

# define function that finds the value of the integral C(v) using simpsons integration
def C(v):
    h = (v - a) / (N - 1)
    x = np.arange(0, v + h ,h)
    C_array = []
    
    for i in range (0, len(x)):
        C_array.append(c(x[i]))
    
    return sp.simps(C_array, x, h)

# define function that finds the value of the integral S(v) using simpsons integration
def S(v):
    h = (v - a) / (N - 1)
    x = np.arange(0, v + h ,h)
    S_array = []
    
    for j in range (0, len(x)):
        S_array.append(s(x[j]))
    
    return sp.simps(S_array, x, h)

# define function that calculates the value of I/I_0 at v using the values of C(v) and S(v) 
def intensity(v):  
    return (1/2) * ((C(v) + 0.5)**2 + (S(v) + 0.5)**2)

# print a value of I/I_0 evaluated at some point (as an example)
print (intensity(10))
```

## Excercise 5

Consider a simple pendulum of length $l$. It oscillates with a maximum angle from the verticle of $\theta_m$. We know from introductory phyicss that if $\theta_m$ is small, the pendulum undergoes simple harmonic motion with period $T_0 = 2 \pi \sqrt{l/g}$ where $g$ is the acceleratoin due to gravity. In the rest of this excercise, we investigate how the period chagnes when the amplitude is no longer small.

### Part a:

Use conservation of energy, with

$$ E = \frac{1}{2} m (l \dot{\theta})^2 + mgl(1-\cos{\theta}) $$

to show that the period of oscillation $T(\theta_m)$ is related to the period of small oscillation $T_0$ through the following integral:


$$ \frac{T(\theta_m)}{T_0} = \frac{\sqrt{2}}{\pi} \int_0^{\theta_m} \frac{d \theta}{\sqrt{\cos(\theta) - \cos(\theta_m)}} $$

#### Proof:

Note that at $\theta = \theta_m$, we have $\dot{\theta} = 0$ so by conservation of energy:

$ E = mgl(1-\cos{\theta_m}) = \frac{1}{2}m(l \dot{\theta})^2 + mgl(1-\cos{\theta}) $

$ \implies mgl(\cos{\theta} - \cos{\theta_m}) = \frac{1}{2} m (l \dot{\theta})^2 $

$ \implies \dot{\theta} = \frac{\sqrt{2gl(\cos\theta - \cos\theta_m)}}{1} $

$ \implies \frac{d \theta}{dt} = \sqrt{\frac{2g}{l}} \sqrt{\cos \theta - \cos \theta_m}$

$ \implies \frac{dt}{d \theta} = \frac{1}{\sqrt{\cos \theta - \cos \theta_m}} $

Note also that $T(\theta) = 4 \int_0^{\theta_m} dt = 4 \int_0^{\theta_m} \frac{dt}{d \theta} d\theta $

So plugging in the equation for $\frac{dt}{d \theta}$ above we find:

$ T(\theta) = 4 \sqrt{\frac{l}{2g}}\int_0^{\theta_m}\frac{1}{\sqrt{\cos \theta - \cos \theta_m}} d\theta $

Which, by recalling that $T_0 = 2 \pi \sqrt(l/g)$ tells us that

$ \frac{T(\theta_m)}{T_0} = \frac{\sqrt{2}}{\pi} \int_0^{\theta_m} \frac{d \theta}{\sqrt{\cos(\theta) - \cos(\theta_m)}} $

QED

### Part b:

Show that we can rewrite the integral above as

$$ \frac{T(\theta_m)}{T_0} = \frac{2}{\pi} \int_0^{\pi/2} \frac{d \psi}{\sqrt{1 - \sin^2(\theta_m / 2)\sin^2(\psi)}}$$

#### Proof

Using $\cos(\theta) = 1-2\sin^2(\theta/2)$ we find

$\frac{T(\theta_m)}{T_0} = \frac{1}{\pi} \int_0^{\theta_m} \frac{d \theta}{\sqrt{\sin^2(\theta_m / 2) - \sin^2(\theta/2)}} $

Further, we claim that $\sin(\psi) = \frac{\sin(\theta / 2)}{\sin (\theta_m / 2)}$ which implies that $d \theta = \frac{2\cos(\psi)\sin(\theta_m/2)}{cos(\theta/2)} d\psi$

plugging this into the integral above and performing the appropriate change in in the limits of te integral we find

$\frac{T(\theta_m)}{T_0} = \frac{1}{\pi} \int_0^{\pi/2} \frac{2\cos(\psi)\sin(\theta_m / 2)}{\cos(\theta/2)\sqrt{\sin^2(\theta_m / 2) - \sin^2(\theta / 2)}} d \psi$

$= \frac{1}{\pi} \int_0^{\pi/2} \frac{2\cos(\psi)}{\cos(\theta/2)\sqrt{1 - \frac{\sin^2(\theta/2)}{\sin^2(\theta_m / 2}}} d \psi$

$= \frac{1}{\pi} \int_0^{\pi/2} \frac{2cos(\psi)}{\cos(\theta/2)\cos(\psi)} d \psi$

$= \frac{2}{\pi} \int_0^{\pi/2} \frac{d \psi}{\cos(\theta / 2)}$

$= \frac{2}{\pi} \int_0^{\pi/2} \frac{d \psi}{\sqrt{1-\sin^2(\theta/2)}}$

$= \frac{2}{\pi} \int_0^{\pi/2} \frac{d \psi}{\sqrt{1 - \sin^2(\theta_m / 2)\sin^2(\psi)}}$

QED
